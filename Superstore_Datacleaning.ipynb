```python
import pandas as pd
```

```python
#Reading the csv file and cleaning the data by droppig the duplicates
df=pd.read_csv("superstore_final_dataset.csv")
df.drop_duplicates(subset=df.columns.difference(['Row_ID']), inplace=True)
```

```python
df.head()
```

```python
#dropping the null values
df.dropna(inplace=True)
df
```

```python
#Checking the order date and ship date columns have nulls in them
#validating and Considering the data with orders ship date> order date
df = df[(pd.to_datetime(df['Order_Date'], errors='coerce').notnull()) &
        (pd.to_datetime(df['Ship_Date'], errors='coerce').notnull()) &
        (pd.to_datetime(df['Ship_Date'], errors='coerce') >= pd.to_datetime(df['Order_Date'], errors='coerce'))]
df = df[['Order_ID', 'Order_Date', 'Ship_Date', 'Ship_Mode', 'Customer_ID', 'Customer_Name','Segment', 'Country', 'City', 'State', 'Postal_Code', 'Region', 'Product_ID', 'Category', 'Sub_Category', 'Product_Name', 'Sales']]
```

```python
#downloading the cleaned csv file
df.to_csv("cleaned_superstore_data.csv", index=False, encoding='utf-8')
```
